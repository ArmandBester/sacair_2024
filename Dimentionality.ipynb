```python
from numpy import random
import pandas as pd
import numpy as np
import scipy
import matplotlib.pyplot as plt

import plotly.express as px

from sklearn.datasets import make_blobs
```

```python
dist_n_features = []
for f in range(1,500):
    X, y_true = make_blobs(n_samples=300, n_features=f, random_state=77, centers=1)
    dists = scipy.spatial.distance.pdist(X)
    min = np.min(dists)
    mean = np.mean(dists)
    max = np.max(dists)
    dist_n_features.append([f, min, max, mean])
dist_n_features = np.array(dist_n_features)
```

```python
plt.figure(figsize=(4,3))
plt.scatter(dist_n_features[:,0], dist_n_features[:,1], label='min', s=1)
plt.scatter(dist_n_features[:,0], dist_n_features[:,2], label='max', s=1)
plt.scatter(dist_n_features[:,0], dist_n_features[:,3], label='mean', s=1)
plt.legend();
```

```python
np.random.seed(42)
x = np.arange(1, 20)
y = np.random.normal(loc=0, scale=3, size=19) + np.arange(11, 30)
z = np.random.normal(loc=0, scale=7, size=19) + np.arange(-20, -1)

fig = plt.figure(figsize=plt.figaspect(0.4))
gs = fig.add_gridspec(1,3,width_ratios=[1,1,2])
ax = fig.add_subplot(gs[0])
ax.scatter(x, np.repeat(0, 19))
ax = fig.add_subplot(gs[1])
ax.scatter(x,y)
ax = fig.add_subplot(gs[2], projection='3d')
ax.scatter(x,y,z)
plt.tight_layout()
```

```python
fig = px.line_3d(x=x,y=y,z=z, height=600, width=800)
fig.update_traces(mode='markers+lines')
fig.update_traces(marker_size = 3)
fig.show()
```
